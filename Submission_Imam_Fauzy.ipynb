```python
# Menyimpan path dari direktori utama yang berisi dataset yang telah dibagi
base_dir = '/tmp/rockpaperscissors/rockpaperscissors/rps-cv-images/split'

# Menyimpan path untuk direktori train yang akan digunakan untuk menyimpan data training set
train_dir = os.path.join(base_dir, 'train')

# Menyimpan path untuk direktori val yang akan digunakan untuk menyimpan data validation set
validation_dir = os.path.join(base_dir, 'val')

# Mengecek isi dari direktori 'train' dan menampilkan list file/folder di dalamnya
os.listdir('/tmp/rockpaperscissors/rockpaperscissors/rps-cv-images/split/train')

# Mengecek isi dari direktori 'val' dan menampilkan list file/folder di dalamnya
os.listdir('/tmp/rockpaperscissors/rockpaperscissors/rps-cv-images/split/val')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Membuat objek ImageDataGenerator untuk augmentasi data pada data latih
train_datagen = ImageDataGenerator(
    rescale=1./255,               # Menykalakan nilai piksel menjadi rentang [0,1]
    rotation_range=40,            # Rentang rotasi gambar (dalam derajat)
    width_shift_range=0.2,        # Rentang pergeseran lebar gambar
    height_shift_range=0.2,       # Rentang pergeseran tinggi gambar
    horizontal_flip=True,         # Melakukan flipping horizontal gambar
    shear_range=0.3,              # Rentang shear (geser) gambar
    zoom_range=0.2,               # Rentang zoom gambar
    fill_mode='nearest'           # Metode pengisian piksel yang hilang
)

# Membuat generator untuk data latih dari direktori
train_generator = train_datagen.flow_from_directory(
    train_dir,                    # Direktori data latih
    target_size=(150, 150),        # Ukuran target gambar
    batch_size=32,                 # Ukuran batch
    class_mode='categorical'       # Mode kelas untuk data kategorikal
)

# Membuat objek ImageDataGenerator untuk normalisasi data pada data validasi
validation_datagen = ImageDataGenerator(rescale=1./255)

# Membuat generator untuk data validasi dari direktori
validation_generator = validation_datagen.flow_from_directory(
    validation_dir,               # Direktori data validasi
    target_size=(150, 150),        # Ukuran target gambar
    batch_size=32,                 # Ukuran batch
    class_mode='categorical'       # Mode kelas untuk data kategorikal
)
```

```python
# Menampilkan indeks kelas pada data training
train_generator.class_indices
```

```python
# Menampilkan indeks kelas pada data validation
validation_generator.class_indices
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

# Membuat objek model sequential untuk membangun arsitektur model
model = tf.keras.models.Sequential([
    # Layer konvolusi pertama dengan 32 filter (3x3), fungsi aktivasi ReLU, dan input gambar 150x150 dengan 3 channel RGB
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),

    # Layer pooling untuk mereduksi dimensi spasial gambar
    tf.keras.layers.MaxPooling2D(2, 2),

    # Layer konvolusi kedua dengan 64 filter (3x3) dan fungsi aktivasi ReLU
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),

    # Layer pooling kedua
    tf.keras.layers.MaxPooling2D(2, 2),

    # Layer konvolusi ketiga dengan 128 filter (3x3) dan fungsi aktivasi ReLU
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),

    # Layer pooling ketiga
    tf.keras.layers.MaxPooling2D(2, 2),

    # Layer konvolusi keempat dengan 512 filter (3x3) dan fungsi aktivasi ReLU
    tf.keras.layers.Conv2D(512, (3, 3), activation='relu'),

    # Layer pooling keempat
    tf.keras.layers.MaxPooling2D(2, 2),

    # Meratakan output menjadi vektor satu dimensi
    tf.keras.layers.Flatten(),

    # Layer fully connected dengan 512 neuron, fungsi aktivasi ReLU, dan regularisasi L2
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),

    # Layer dropout untuk mengurangi overfitting
    tf.keras.layers.Dropout(0.5),

    # Layer output dengan 3 neuron dan fungsi aktivasi softmax (sesuai dengan jumlah kelas)
    tf.keras.layers.Dense(3, activation='softmax')
])

# Menampilkan ringkasan arsitektur model
model.summary()

# Callback EarlyStopping untuk menghentikan pelatihan jika tidak ada peningkatan dalam metrik yang dipantau (biasanya validasi loss)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Callback ReduceLROnPlateau untuk mengurangi laju pembelajaran jika tidak ada peningkatan dalam metrik yang dipantau
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)

# Optimizer Adam untuk pelatihan model
optimizer = Adam(learning_rate=0.001)

# Mengkompilasi model dengan fungsi categorical_crossentropy dan optimizer Adam
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
# Mengkompilasi model dengan menggunakan kategori crossentropy sebagai fungsi kerugian, optimizer Adam dengan laju pembelajaran 0.0001, dan metrik akurasi
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=0.0001),
              metrics=['accuracy'])

# Callback EarlyStopping untuk menghentikan pelatihan jika tidak ada peningkatan dalam validasi loss selama 5 epo
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Callback ReduceLROnPlateau untuk mengurangi laju pembelajaran jika tidak ada peningkatan dalam validasi loss selama 3 epo, dengan faktor pengurangan 0.2 dan batas bawah laju pembelajaran 1e-6
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)

# Pelatihan model dengan data generator untuk data latih dan data validasi, menggunakan callback EarlyStopping dan ReduceLROnPlateau, selama 20 epo
history = model.fit(
        train_generator,
        steps_per_epoch=len(train_generator),
        epochs=22,
        validation_data=validation_generator,
        validation_steps=len(validation_generator),
        callbacks=[early_stopping, reduce_lr],
        verbose=2
)
```

```python
# Evaluasi model pada data pelatihan
score_train = model.evaluate(train_generator)

# Menampilkan hasil evaluasi pada data pelatihan
print('Loss (Train): {:.5f}'.format(score_train[0]))
print('Accuracy (Train): {:.5f}'.format(score_train[1]))
```

```python
# Evaluasi model pada data validasi
score_validation = model.evaluate(validation_generator)

# Menampilkan hasil evaluasi pada data validasi
print('Loss (Validation): {:.5f}'.format(score_validation[0]))
print('Accuracy (Validation): {:.5f}'.format(score_validation[1]))
```
